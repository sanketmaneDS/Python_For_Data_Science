# Matplotlib

```python
#install and then import matplotlib
import matplotlib.pyplot as plt
```

```python
#matplotlib to render plots in the notebook
%matplotlib inline
```

```python
x = [-1, 2, 5, 8, 10]
y = [10, 3, -4, 6, 2]
```

```python
a = plt.figure(figsize=(15,3))   # figsize(width, height) in inches
# figure of dimensions specified above will be stored in variable named "a"

plt.plot(x, y)                      
plt.xlim(0, 10)  # range of X is set to 0 to 10, points beyond this range will not be plotted
plt.ylim(-3, 8)  # range of Y is set to -3 to 8, points beyond this range will not be plotted

plt.xlabel('X Axis')
plt.ylabel('Y axis')
plt.title('Line Plot')
plt.suptitle('Sales Comaprison', size=18, y=1.05)   # super title; y is distance between title and suptitle
```

# Matplotlib

```python
plt.plot(mtcars.hp)  #line plot-> when you need to visualise data over a period of time
plt.show()
```

```python
plt.scatter(mtcars.mpg,mtcars.disp)  #scatter plot-> to determine if any co relation exist b/n 2 variables

plt.xlabel("mpg")
plt.ylabel("disp")
plt.title("mpg vs disp: negative linear association")   

plt.show()
```

```python
#creates histogram of 5 bins with colours specified in command

plt.hist(mtcars['mpg'],facecolor ="orange",edgecolor ="red",bins =5)
plt.show()
```

```python
#Boxplot
#help(plt.boxplot) for additional info

plt.boxplot(mtcars['disp']) # by default, boxplot is vertical
plt.ylabel("displacement")
plt.xlabel("")
plt.title("Boxplot")  
plt.show()
```

```python
plt.boxplot(mtcars['disp'],vert =False) # horizontal boxplot
plt.show()
```

```python
plt.subplot(1,2,1)
plt.boxplot(mtcars['disp'])

plt.subplot(1,2,2)
plt.boxplot(mtcars['disp'],vert=False)

plt.show()
```

```python
#alternative way of subplotting
fig, (ax1, ax2) = plt.subplots(1,2)

ax1.boxplot(mtcars['disp'])
ax2.boxplot(mtcars["disp"],vert=False)

plt.show()
```

```python
#Violin Plot
#help(plt.violinplot) for additional info

plt.violinplot(mtcars["disp"])
plt.show()
```

```python
plt.subplot(1,2,1)
plt.boxplot(mtcars['disp'])

plt.subplot(1,2,2)
plt.violinplot(mtcars["disp"])

plt.show() 

#While a box plot only shows summary statistics such as mean/median and interquartile ranges, the violin plot shows the full distribution of the data
```

## Univariate  one variable data visualization

```python
tips =sns.load_dataset('tips')
```

```python
tips
```

```python
#strip plot

sns.stripplot(y= 'tip', data=tips, jitter=True) #Jitter make overlapping markers in a scatter plot visible.
plt.ylabel('tip ($)')
plt.show()
```

```python
#Grouping with stripplot()
sns.stripplot(x='time', y='tip', data=tips, jitter=False)
plt.ylabel('tip ($)')
plt.show()
```

```python
sns.stripplot(x='day', y='tip', data=tips,size=4,jitter=True)
plt.ylabel('tip ($)')
plt.show()
```

```python
#Swarm plot
#data points do not overlap and are adjusted along the categorical axis.

sns.swarmplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
    
```

```python
sns.swarmplot(x='day', y='tip', data=tips, hue='time')
plt.ylabel('tip ($)')
plt.show()
```

```python
#Changing orientation
sns.swarmplot(x='tip', y='day', data=tips, hue='sex',orient='h')
plt.xlabel('tip ($)')
plt.show()
```

```python
#Box plot

sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')

plt.show()
```

```python
#Violin plot

sns.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')

plt.show()
```

```python
plt.subplot(1,2,1)
sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')

plt.subplot(1,2,2)
sns.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')

plt.show()
```

```python
#Combining plots

sns.violinplot(x='day', y='tip', data=tips,inner=None, color='lightgray')
sns.stripplot(x='day', y='tip', data=tips, size=4,jitter=True)
plt.ylabel('tip ($)')
plt.show()
```

## Bivariate  two variables data visualization

```python
#Joint plot

sns.jointplot(x= 'total_bill', y= 'tip', data=tips)
plt.show()
```

```python
# Density plot
sns.jointplot(x='total_bill', y= 'tip', data=tips,kind='kde')
plt.show()
```

```python
#Pair plot

sns.pairplot(tips)
plt.show()
```
